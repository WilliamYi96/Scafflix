```python
# Fundamental environment setup
import matplotlib
import numpy as np
import psutil
import seaborn as sns
import copy
import urllib.request
import sklearn.datasets
import matplotlib.pyplot as plt
import numpy.linalg as la
from scipy.sparse import csc_matrix, csr_matrix
from sklearn.datasets import load_svmlight_file, fetch_rcv1

sns.set(style="whitegrid", context="talk", font_scale=1.2, palette=sns.color_palette("bright"), color_codes=False)
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['font.sans-serif'] = 'DejaVu Sans'
matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['figure.figsize'] = (9, 6)
```

```python
p = 0.2
max_rounds = 1000

scafflixs = []

file_numbers = ['0001', '0002', '0003', '0004', '0005']

for file_number in file_numbers:
    file_name = f'../logs/v1_30points/{file_number}.txt'
    with open(file_name, 'r') as minput:
        _ = minput.readline().split(',')
        _ = minput.readline().split(',')
        cur_scafflix = minput.readline().split(',')
        cur_scafflix = [float(i) for i in cur_scafflix]
        scafflixs.append(cur_scafflix)

print(scafflixs[0][:10])
print(scafflixs[1][:10])
mlen = 100
round_nums = np.linspace(int(max_rounds / mlen), max_rounds, num=mlen, endpoint=True)
# round_nums = np.arange(int(max_rounds / mlen), max_rounds + int(max_rounds / mlen), int(max_rounds / mlen))

interval = int(1 / p) 
for i in range(len(scafflixs)):
    scafflixs[i] = scafflixs[i][0:len(scafflixs[i]):int(len(scafflixs[i])/mlen)]


size = 30
markevery = 1.5
plot_nums = 500
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['xtick.labelsize'] = 24  # 40
plt.rcParams['ytick.labelsize'] = 24  # 40
plt.rcParams['legend.fontsize'] = 20  # 30
plt.rcParams['axes.titlesize'] = 28  # 40
plt.rcParams['axes.labelsize'] = 28  # 40
plt.rcParams["figure.figsize"] = [13, 9]
plt.ylabel('test accuracy')
plt.xlabel('communication rounds')
# plt.yticks(range(0,1))
# plt.yscale('log')
# plt.xscale('log')
plt.title(fr"FEMNIST")
# plt.legend()
plt.grid(visible=True)
markevery = 100
# plt.yticks(np.arange(0, 1, 0.2))

start, end = 2, plot_nums - 1
plt.plot(round_nums[start:end], scafflixs[0][start:end], label=fr'Scafflix, $\alpha=0.1$')
plt.plot(round_nums[start:end], scafflixs[1][start:end], label=fr'Scafflix, $\alpha=0.3$')
plt.plot(round_nums[start:end], scafflixs[2][start:end], label=fr'Scafflix, $\alpha=0.5$')
plt.plot(round_nums[start:end], scafflixs[3][start:end], label=fr'Scafflix, $\alpha=0.7$')
plt.plot(round_nums[start:end], scafflixs[4][start:end], label=fr'Scafflix, $\alpha=0.9$')

plt.legend()
plt.savefig(f'0301.pdf')
plt.show()
plt.close()
```
